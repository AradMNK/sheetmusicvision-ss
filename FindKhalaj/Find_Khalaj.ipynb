```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("background.JPG")
img_grayscale = cv.imread("background.JPG", cv.IMREAD_GRAYSCALE)

template = cv.imread("template.JPG", cv.IMREAD_GRAYSCALE)

w, h = template.shape[::-1]

method = 'cv.TM_CCOEFF'
method = eval(method)

res = cv.matchTemplate(img_grayscale, template, method)
min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)

top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

cv.rectangle(img,top_left, bottom_right, (0, 255, 0), 3)

img2 = img[:,:, (2, 1, 0)]

_ = plt.imshow(img2)
_ = plt.xticks([])
_ = plt.yticks([])
_ = plt.box(0)

_ = cv.imwrite("TemplateMatching.JPG", img)
```
